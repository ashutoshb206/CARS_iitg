### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# Calculate horsepower-to-weight ratio
data['hp_to_weight'] = data['horsepower'] / data['weight']

# Find median mpg
median_mpg = data['mpg'].median()

# Filter cars with mpg > median
above_median = data[data['mpg'] > median_mpg]

# Find car(s) with best ratio
best_ratio = above_median['hp_to_weight'].max()
best_cars = above_median[above_median['hp_to_weight'] == best_ratio]
print("Car(s) with best hp/weight ratio and above-median mpg:")
print(best_cars[['mpg', 'horsepower', 'weight', 'hp_to_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
# Group and pivot
mpg_by_year_origin = data.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()
pivot_table = mpg_by_year_origin.pivot(index='model_year', columns='origin', values='mpg')

# Plot
pivot_table.plot(marker='o', figsize=(10, 6))
plt.title("Average MPG Over Years by Origin")
plt.xlabel("Model Year")
plt.ylabel("Average MPG")
plt.grid(True)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
#your code here
# Reset index to access 'name' as a column
data_reset = data.reset_index()

# Group by name
grouped = data_reset.groupby('name')

# Filter for cars that appear in multiple years with low std in mpg
consistent = grouped.filter(lambda x: x['model_year'].nunique() >= 2 and x['mpg'].std() < 1.0)

# Group again to summarize
summary = consistent.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    avg_mpg=('mpg', 'mean'),
    std_mpg=('mpg', 'std')
)

# Final filter and sort
final_result = summary[summary['std_mpg'] < 1.0].sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])
print("Consistent car models:")
print(final_result[['appearances', 'avg_mpg']])
```
